```python
import numpy as np
import matplotlib.pyplot as plt
```

# data

```python
x = np.random.rand(20) * 8 - 4
x
```

```python
y = np.sin(x) + np.random.rand(20) * 0.2
y
```

# 1st trial of modeling to fit data

```python
omega = np.polyfit(x, y, 1)
omega
```

# evaluate the model

```python
f = np.poly1d(omega)
f
```

```python
plt.xlabel('X')
plt.ylabel('y')
plt.title('using polyfitfunction')
plt.grid()
plt.scatter(x, y, marker='x', c='red')
x_axis = np.linspace(-4, 4, 100)
plt.plot(x_axis, f(x_axis), color='green')
plt.show()
```

The 1st round linear model sucks, change to quadratic equation to fit the data

# 2nd trial of modeling to fit data

```python
omega = np.polyfit(x, y, 2)
omega
```

# evaluate the model

```python
f = np.poly1d(omega)
f
```

```python
def plot_fitting_curve(f, x, y):
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title('using polyfitfunction')
    plt.grid()
    plt.scatter(x, y, marker='x', c='red')
    x_axis = np.linspace(-4, 4, 100)
    plt.plot(x_axis, f(x_axis), color='green')
    plt.show()

plot_fitting_curve(f, x, y)
```

looks better but there's still a room for improvement 

# 3rd trial of modeling to fit data

```python
omega = np.polyfit(x, y, 3)
omega
```

# evaluate the model

```python
f = np.poly1d(omega)
f
```

```python
plot_fitting_curve(f, x, y)
```

looks much better, see if we could make the curve to the perfection

# 4th trial of modeling to fit data

```python
omega = np.polyfit(x, y, 4)
omega
```

# evaluate the model

```python
f = np.poly1d(omega)
f
```

```python
plot_fitting_curve(f, x, y)
```

# 5th trial of modeling to fit data

```python
omega = np.polyfit(x, y, 5)
omega
```

# evaluate the model

```python
f = np.poly1d(omega)
f
```

```python
plot_fitting_curve(f, x, y)
```

The 5th round of trial seems overfitting, so the cubic model or quartic model could be a better choice.
