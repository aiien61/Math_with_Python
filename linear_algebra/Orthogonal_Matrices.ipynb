```python
import numpy as np
import torch
import tensorflow as tf
```

```python
I = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
I
```

```python
col_1 = I[:, 0]
col_1
```

```python
col_2 = I[:, 1]
col_2
```

```python
col_3 = I[:, 2]
col_3
```

Use dot product `np.dot()` method to prove each column of $I_3$ is orthogonal to each other.

```python
np.dot(col_1, col_2)
```

```python
np.dot(col_1, col_3)
```

```python
np.dot(col_2, col_3)
```

Use `np.linalg.norm()` method to prove each column of $I_3$ has unit norm.

```python
np.linalg.norm(col_1)
```

```python
np.linalg.norm(col_2)
```

```python
np.linalg.norm(col_3)
```

Since the matrix $I_3$ has mutually orthogonal columns and each column has unit norm, the column vectors of $I_3$ are orthonormal.

Also, since $I_3^T=I_3$, this means that rows of $I_3$ are orthonormal, $I_3$ is an orthogonal matrix.

```python
K = torch.tensor([ [2/3, 1/3, 2/3], [-2/3, 2/3, 1/3], [1/3, 2/3, -2/3] ])
K
```

```python
K_col1 = K[:, 0]
K_col1
```

```python
K_col2 = K[:, 1]
K_col2
```

```python
K_col3 = K[:, 2]
K_col3
```

Prove columns of matrix K are mutually orthogonal

```python
torch.dot(K_col1, K_col2)
```

```python
torch.dot(K_col1, K_col3)
```

```python
torch.dot(K_col3, K_col2)
```

Prove that every column of $K$ has unit norm

```python
torch.norm(K_col1)
```

```python
torch.norm(K_col2)
```

```python
torch.norm(K_col3)
```

Determined that, in addition to being orthogonal, the columns of $K$ has unit norm, they therefore are orthonormal.

```python
torch.matmul(K.T, K)
```

$K^TK=I$ has been confirmed and therefore $K$ is an orthogonal matrix.
