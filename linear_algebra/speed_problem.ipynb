```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# points for drawing
t = np.linspace(0, 40, 1000) # start, stop, n points
```

Distance travelled by robber: $d = 2.5t$

```python
dist_robber = 2.5 * t
```

Distance travelled by sheriff: $d = 3(t - 5)$

```python
dist_sheriff = 3 * (t - 5)
```

```python
fig, ax = plt.subplots()
plt.title('The time when a bank robber is being caught')
plt.xlabel('time (in minutes)')
plt.ylabel('distance (in km)')
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, dist_robber, color='red')
ax.plot(t, dist_sheriff, color='blue')
plt.axvline(x=30, color='purple', linestyle='--')
_ = plt.axhline(y=75, color='purple', linestyle='--')
```

On April 1st, solar panel "Mark I" design had begun generating power: 1kj/day.

On May 1st, solar panel "Mark II" desing had begun generating power: 4 kj/day.

```python
t = np.linspace(0, 20, 1000)
```

```python
p_m1 = 1 * (t + 30)
p_m2 = 4 * t
```

What day is it when Mark II design has generated as much total energy as the Mark I design?

How much total energy have both generated by that day?

```python
fig, ax = plt.subplots()
plt.title('Total Power Generated by Both Designs')
plt.xlabel('time (in days)')
plt.ylabel('energy (kj)')
ax.set_xlim([0, 20])
ax.set_ylim([0, 100])
ax.plot(t, p_m1, color='green')
ax.plot(t, p_m2, color='red')
plt.axvline(x=10, color='purple', linestyle='--')
_ = plt.axhline(y=40, color='purple', linestyle='--')
```
