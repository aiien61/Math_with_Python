```python
import numpy as np
import torch
import tensorflow as tf
```

```python
X = np.array([[1,2], [3,4], [5,6]])
X_pt = torch.tensor([[1,2], [3,4], [5,6]])
X_tf = tf.Variable([[1,2], [3,4], [5,6]])
```

```python
X
```

```python
X.sum()
```

```python
torch.sum(X_pt)
```

```python
tf.reduce_sum(X_tf)
```

Can also be done one specific axis alone

```python
# summing over all rows
X.sum(axis=0)
```

```python
# summing over all columns
X.sum(axis=1)
```

```python
torch.sum(X_pt, 0)
```

```python
tf.reduce_sum(X_tf, 1)
```

Other operations that can be applied with reduction along all or a selection of axes:

*   maximum
*   minimum
*   mean
*   product
