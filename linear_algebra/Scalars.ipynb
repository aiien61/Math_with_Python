Scalars in TensorFlow

```python
import tensorflow as tf
```

```python
x = tf.Variable(25, dtype=tf.int16) # dtype is optional
x
```

```python
x.shape
```

```python
y = tf.Variable(3, dtype=tf.int16)
y
```

```python
tf_sum = tf.add(x, y)
tf_sum
```

```python
# .numpy() operations automatically convert tensors to NumPy arrays, and vice versa.
tf_sum.numpy()
```

```python
type(tf_sum.numpy())
```

```python
tf_float = tf.Variable(25, dtype=tf.float32)
tf_float
```
