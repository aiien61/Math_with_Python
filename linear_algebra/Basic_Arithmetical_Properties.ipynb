```python
import numpy as np
import torch
import tensorflow as tf
```

```python
X = np.array([[1, 2], [3, 4], [5, 6]])
X_pt = torch.from_numpy(X)
X_tf = tf.Variable([[1, 2], [3, 4], [5, 6]])
```

```python
X
```

```python
X * 2
```

```python
X + 2
```

```python
X * 2 + 2
```

```python
X_pt * 2 + 2 # Python operators are overloaded, but allowed alternatively to use torch.mul() for multiplication or torch.add() for addition
```

```python
torch.add(torch.mul(X_pt, 2), 2)
```

```python
X_tf * 2 + 2 # Operators likewise overloaded, and allowed alternatively to use tf.multiply() for multiplication or tf.add() for addition
```

```python
tf.add(tf.multiply(X_tf, 2), 2)
```
