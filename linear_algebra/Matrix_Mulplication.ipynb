```python
import numpy as np
import torch
import tensorflow as tf
```

Matrix Multiplication (with Vector)

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
A
```

```python
b = np.array([1, 2])
b
```

```python
np.matmul(A, b)
```

```python
np.dot(A, b)
```

```python
np.all(np.matmul(A, b) == np.dot(A, b))
```

```python
A_pt = torch.tensor([[1, 2], [3, 4], [5, 6]])
A_pt
```

```python
b_pt = torch.tensor([1, 2])
b_pt
```

```python
torch.matmul(A_pt, b_pt)
```

```python
A_tf = tf.Variable([[1, 2], [3, 4], [5, 6]])
A_tf
```

```python
b_tf = tf.Variable([1, 2])
b_tf
```

```python
tf.linalg.matvec(A_tf, b_tf)
```

```python
B = np.array([[1, 0], [0, 1]])
B
```

```python
np.dot(A, B)
```

```python
np.all(np.matmul(A, B) == np.dot(A, B))
```

```python
torch.matmul(A_pt, B_pt)
```

```python
B_tf = tf.convert_to_tensor(B, dtype=tf.int32)
B_tf
```

```python
tf.matmul(A_tf, B_tf)
```
