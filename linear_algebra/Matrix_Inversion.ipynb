####Geometric Visualisation
Two equations in the system are:

$4b+2c=4$

$-5b-3c=-7$

Both equations cab be rearranged to isolate a variable, say _c_, Starting with the first equation:

$4b+2c=4$

$2b+c=2$

$c=2-2b$

Then for the second equation:

$-5b-3c=-7$

$-3c=-7+5b$

$c=\frac{-7+5b}{-3}=\frac{7-5b}{3}$

```python
b = np.linspace(-10, 10, 1000)
```

```python
c1 = 2 - 2 * b
c2 = (7 - 5 * b) / 3
```

```python
fig, ax = plt.subplots()

plt.xlabel('b', c='darkorange')
plt.ylabel('c', c='brown')

plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

ax.set_xlim([-2, 3])
ax.set_ylim([-1, 5])
ax.plot(b, c1, c='purple')
ax.plot(b, c2, c='purple')

plt.axvline(x=-1, color='green', linestyle='--')
_ = plt.axhline(y=4, color='green', linestyle='--')
```

###Matrix Inversion Where No Solution

*   Matrix is singular
*   Some columns are linearly dependent





```python
X = np.array([[1, 2], [2, 4]])
X
```

```python
from numpy.linalg.linalg import LinAlgError

try:
  X_inv = np.linalg.inv(X)
except LinAlgError as e:
  print(f"X has no inverse because: {e}")
```
