Vector Transposition

```python
# Transposing a regular 1-D array has no effect.
x_t = x.T
x_t
```

```python
x_t.shape
```

```python
# But it does when we use nested "matrix-style" brackets
y = np.array([[18, 3, 6]])
y
```

```python
y.shape
```

```python
y_t = y.T
y_t
```

```python
y_t.shape
```

```python
y_t.T
```

```python
y_t.T.shape
```
